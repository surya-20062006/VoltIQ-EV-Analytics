# Dataset Understanding

- Dataset contains 96,845 records.
- No missing values are present.
- No duplicate records are found.
- Data covers EV sales from 2014 to 2024.
- Dataset contains 34 states.
- There are 5 vehicle categories and 12 vehicle types.
- EV sales data is highly skewed with many zero values.
- Data type optimization is required before analysis.

# Complete data cleaning and preprocessing

```python
df_clean=df.copy()
```

```python
df_clean.dtypes
```

```python
df_clean["Year"] = df_clean["Year"].astype(int)
```

```python
df_clean["Date"] = pd.to_datetime(df_clean["Date"])
```

```python
df_clean["EV_Sales_Quantity"]=df_clean["EV_Sales_Quantity"].astype(int)
```

```python
df_clean.info()
```

```python
df_clean["Month_Number"] = df_clean["Date"].dt.month
```

```python
df_clean["Quarter"] = df_clean["Date"].dt.quarter
```

```python
df_clean.head()
```

```python
df_clean.to_csv(
    "../data/proceed/cleaned_ev_sales.csv",
    index=False
)
```

## Operations Performed

- Created a copy of the original dataset.
- Converted Year from float to integer.
- Converted EV Sales Quantity from float to integer.
- Converted Date to datetime format.
- Created Month_Number feature.
- Created Quarter feature.
- Saved the cleaned dataset.

# Exploratory Data Analysis (EDA)

```python
df_clean["EV_Sales_Quantity"].sum()
```

```python
year_sales = df_clean.groupby("Year")["EV_Sales_Quantity"].sum()
year_sales
```

```python
import matplotlib.pyplot as plt

year_sales.plot(kind="bar", figsize=(10,5))
plt.title("Year-wise EV Sales")
plt.xlabel("Year")
plt.ylabel("EV Sales")
plt.show()
```

```python
top_states = (
    df_clean.groupby("State")["EV_Sales_Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_states
```

```python
top_states.plot(kind="barh", figsize=(10,5))
plt.title("Top 10 EV States")
plt.show()
```

```python
category_sales = (
    df_clean.groupby("Vehicle_Category")
    ["EV_Sales_Quantity"]
    .sum()
)

category_sales
```

```python
category_sales.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(8,8)
)
plt.ylabel("")
plt.show()
```

```python
vehicle_type = (
    df_clean.groupby("Vehicle_Type")
    ["EV_Sales_Quantity"]
    .sum()
    .sort_values(ascending=False)
)

vehicle_type
```

```python
monthly_sales = (
    df_clean.groupby("Month_Name")
    ["EV_Sales_Quantity"]
    .sum()
)

monthly_sales
```

```python
vehicle_class = (
    df_clean.groupby("Vehicle_Class")
    ["EV_Sales_Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

vehicle_class
```
